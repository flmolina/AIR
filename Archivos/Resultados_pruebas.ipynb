```python
import os
import numpy as np

# Ruta de la carpeta que contiene los archivos CSV
carpeta = "RESULTADOS"

# Lista para almacenar los vectores y nombres de archivos
vectores = []

# Recorre todos los archivos en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith(".csv"):
        # Lee el archivo CSV línea por línea y convierte cada línea en un elemento del vector numpy
        nombre_archivo = archivo.split(".")[0]  # Obtiene el nombre del archivo sin la extensión
        ruta_completa = os.path.join(carpeta, archivo)
        
        with open(ruta_completa, 'r') as f:
            contenido = f.readlines()
        
        # Elimina los saltos de línea y convierte el contenido en un vector numpy
        datos = np.array([float(linea.strip()) for linea in contenido])
        
        vectores.append((nombre_archivo, datos))
Resultado_IA=vectores[0][1]
Resultado_IAP=vectores[1][1]
Resultado_IB=vectores[2][1]
Resultado_IBP=vectores[3][1]
Resultado_IC=vectores[4][1]
Resultado_VA=vectores[5][1]
Resultado_VAP=vectores[6][1]
Resultado_VB=vectores[7][1]
Resultado_VBP=vectores[8][1]
Resultado_VC=vectores[9][1]
Resultado_VCP=vectores[10][1]
Resultado_velocidad=vectores[11][1]
Resultado_ICP=Resultado_IAP-Resultado_IBP
```

```python
def media_y_varianza(vector):
    media = np.mean(vector)
    varianza = np.var(vector)
    std = np.sqrt(varianza)
    return media,  std

def error_media(vector):
    return (abs(vector-np.mean(vector))/np.mean(vector))*100
Resultado_IA=error_media(Resultado_IA)  
Resultado_IAP=error_media(Resultado_IAP)
Resultado_IB=error_media(Resultado_IB)
Resultado_IBP=error_media(Resultado_IBP)
Resultado_IC=error_media(Resultado_IC)
Resultado_VA=error_media(Resultado_VA)
Resultado_VAP=error_media(Resultado_VAP)
Resultado_VB=error_media(Resultado_VB)
Resultado_VBP=error_media(Resultado_VBP)
Resultado_VC=error_media(Resultado_VC)
Resultado_VCP=error_media(Resultado_VCP)
Resultado_ICP=error_media(Resultado_ICP)

```
